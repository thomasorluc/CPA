```python
import networkx as nx
import random
import matplotlib
```

```python
def generate_graph(N,p,nbC=1,q=0):
    graph = nx.Graph()
    clusterSize = int (N / nbC)
    #print(clusterSize)
    clusterI = 0
    clusterJ = 0
    for i in range(1,N+1):
        clusterI = int ((i-1) / clusterSize)
        for j in range(i+1,N+1):
            clusterJ = int ((j-1) / clusterSize)
            proba = p if clusterI == clusterJ else q
            if(random.random() < proba):
                graph.add_edge(i,j)
    nx.draw(graph)
    return graph
```

```python
p = 0.5
q = p / 1000
graphe = generate_graph(400,p,4,q)
```

```python
f = open("graphePython.txt", "a")
for i in list(graphe.edges()):
    f.write(str(i[0]))
    f.write(" ")
    f.write(str(i[1]))
    f.write("\n")
    
f.close()
```

```python
f = open("graph2t.txt", "r")
graphe2 = nx.Graph()
for x in f:
    y = x.split()
    graphe2.add_edge(int(y[0]),int(y[1]))
f.close()
f = open("graphe2Label.txt", "r")
listeLabels = []
for x in f:
    y = x.split()
    if int(y[0]):
        listeLabels.append([int(y[0]),int(y[1])])
f.close()

print(sum(list(nx.triangles(graphe2).values()))/3)

labels = [ b for (a,b) in listeLabels]
labels = list(set(labels))
print(labels)

listeLabels2 = []
for i in labels:
    listeLabels2.append([])
for i in range(len(labels)):
    for (val,key) in listeLabels:
        if key == labels[i]:
            listeLabels2[i].append(val)
"""print(listeLabels2)"""

pos = nx.spring_layout(graphe2)

nx.draw_networkx_nodes(graphe2,pos,nodelist=listeLabels2[0],node_color='b')
nx.draw_networkx_nodes(graphe2,pos,nodelist=listeLabels2[1],node_color='r')
nx.draw_networkx_nodes(graphe2,pos,nodelist=listeLabels2[2],node_color='g')
nx.draw_networkx_nodes(graphe2,pos,nodelist=listeLabels2[3],node_color='y')

nx.draw_networkx_edges(graphe2,pos)
```
